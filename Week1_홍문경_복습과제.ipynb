## 6-1번) 
sklearn(사이킷런)의 StratifiedShuffleSplit을 이용하여 train set과 test set을 샘플링하시오. 단, test set의 size는 전체 데이터셋의 20%로 한다. 
- 각 set의 이름은 strat_train_set, strat_test_set으로 한다. 

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
```

```python
for train_index,test_index in split.split(titanic, titanic["Fare_cat"]):
    strat_train_set=titanic.loc[train_index]
    strat_test_set=titanic.loc[test_index]
```

# 7. 모델 평가  

## 7-1번)
교차 검증(cross_val_score)을 사용하여 모델 평가하고, RMSE의 평균과 표준편차를 출력하시오.

- `X_train, y_train`을 사용할 것
- 모델: DecisionTreeClassifier
- 예측 성능 지표: accuracy
- fold: 5

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
# 필요없는 변수를 제거하고 모델을 돌리겠습니다
# PassengerID, Name, Cabin
my_X_train=X_train[["PassengerId","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
```

```python
my_X_train # 이 중 명목형 변수인 Sex,Embarked 는 인코딩
```

```python
my_titanic_dum=pd.get_dummies(my_X_train,drop_first=True)
my_titanic_dum
```

```python
tree_class = DecisionTreeClassifier()
tree_class.fit(my_titanic_dum, y_train)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_class, my_titanic_dum, y_train,
                         scoring="accuracy", cv=5)
scores
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_class, my_titanic_dum, y_train,
                         scoring="neg_mean_squared_error", cv=5)
tree_rmse_scores = np.sqrt(-scores)
tree_rmse_scores
```

```python
import numpy as np
print(np.sum(tree_rmse_scores)/len(tree_rmse_scores))
print(np.std(tree_rmse_scores))
```
